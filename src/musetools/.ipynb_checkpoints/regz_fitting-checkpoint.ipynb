```python
from astropy.io import fits
from astropy.io import ascii
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
from musetools import io as io
from musetools import spec as s
import musetools.util as u
import musetools.modeling as m
from astropy.wcs import WCS
from lmfit import Model
import pdb
%matplotlib inline
import getpass
from scipy.optimize import curve_fit
username=getpass.getuser()

if username == 'bordoloi':
	fitsfile = '/Users/bordoloi/Dropbox/MUSE/LensedArc/RCS0327_16mc_zap.fits'
else:
	fitsfile = '/home/ahmed/astro/data/RCS0327_16mc_zap.fits'

import importlib
```

```python
#xcen = [114,114,115,118,121,124,127,133,137,141,148,153,160, 166, 170,177,185,191,198,203,208,213,220,225,231,238,244,246,244,240,238,242]
#ycen = [226,233,237,241,244,248,252,257,260,264,269,271,274, 274, 274,274,274,272,271,270,268,266,263,259,255,249,244,240,237,234,228,224]
z_e = 1.7037455
z_u = 1.7038840
z_b = 1.7036000
z_g = 1.7038500
xcen_e = [114,114,115,141,148]
ycen_e = [226,233,237,264,269]

xcen_u = [118,121,124,127,133,137]
ycen_u = [241,244,248,252,257,260]

xcen_b = [153,160]
ycen_b = [271,274]

xcen_g = [244,246,244,240,238,242]
ycen_g = [244,240,237,234,228,224]


def mg_fit(cx,cy,z,plot=False):
    data = ascii.read('/home/ahmed/astro/spectra/spectrum_mg_'+str(cx)+'_'+str(cy)+'.dat')
    # wave wrest flx flx_er continuum flx_norm flx_er_norm
    wave        = data['wave']
    wrest       = wave / (1. + z)
    flx         = data['flx']
    flx_er      = data['flx_er']
    continuum   = data['continuum']
    flx_norm    = data['flx_norm']
    flx_er_norm = data['flx_er_norm']
    lam_center = [2796.351,2803.528]
    #lam_center = [2586.650,2600.173,2612.654,2626.451]
    vel1 = u.veldiff(wrest,lam_center[0])
    popt, pcov = curve_fit(m.modelMg, vel1, flx_norm,[ 0., 130., 0.9, 1., 0.8, 0.7, 150., 80.])
    perr = np.sqrt(np.diag(pcov))

    #print(popt)
    #print(perr)
        
    Fabs = m.modelMg(vel1,*popt)
    if plot ==True:
        fig, ax = plt.subplots()
        ax.step(vel1, flx_norm, label='Normalized Flux')
        #ax.plot(vel1, m.modelMg(vel1,*popt), 'y-',label='Model')
        ax.plot(vel1, Fabs, label='Absorption')
        ax.step(vel1, flx_er_norm,'r',label='Error')
        ax.legend(loc=0)
        ax.set_title('Normalized Flux Vs Velocity for '+str(cx)+' & '+str(cy)+'')
        ax.set_xlabel('Velocity')
        ax.set_ylabel('Normalized Flux')
        ax.set_xlim([-2100,6200])
        fig.savefig('/home/ahmed/astro/figures/fitting/regions/model_fit_mg_'+str(cx)+'_'+str(cy)+'.pdf',overwrite=True)
        plt.show()
    return popt, perr


def fe_fit(cx,cy,z,plot=False):
    data = ascii.read('/home/ahmed/astro/spectra/spectrum_fe_'+str(cx)+'_'+str(cy)+'.dat')
    # wave wrest flx flx_er continuum flx_norm flx_er_norm
    wave        = data['wave']
    wrest       = wave / (1. + z)
    flx         = data['flx']
    flx_er      = data['flx_er']
    continuum   = data['continuum']
    flx_norm    = data['flx_norm']
    flx_er_norm = data['flx_er_norm']
    #lam_center = [2796.351,2803.528]
    lam_center = [2586.650,2600.173,2612.654,2626.451]
    vel1 = u.veldiff(wrest,lam_center[0])
    popt, pcov = curve_fit(m.modelFe, vel1, flx_norm,[0.,0.7,0.4,1.1,1.7,1.,150.,100.])
    perr = np.sqrt(np.diag(pcov))

    #print(popt)
    #print(perr)
        
    Fabs = m.modelFe(vel1,*popt)
    if plot ==True:
        fig, ax = plt.subplots()
        ax.step(vel1, flx_norm, label='Normalized Flux')
        #ax.plot(vel1, m.modelMg(vel1,*popt), 'y-',label='Model')
        ax.plot(vel1, Fabs, label='Absorption')
        ax.step(vel1, flx_er_norm,'r',label='Error')
        ax.legend(loc=0)
        ax.set_title('Normalized Flux Vs Velocity for '+str(cx)+' & '+str(cy)+'')
        ax.set_xlabel('Velocity')
        ax.set_ylabel('Normalized Flux')
        ax.set_xlim([-2100,6200])
        fig.savefig('/home/ahmed/astro/figures/fitting/regions/model_fit_fe_'+str(cx)+'_'+str(cy)+'.pdf',overwrite=True)
        plt.show()
    return popt, perr
'''
vabs = []
vems = []
vabs_er = []
vems_er = []

vabs.append(popt[0])
vems.append(popt[1])
vabs_er.append(perr[0])
vems_er.append(perr[1])
'''
```

```python
### For the MgII lines in the squares in region E
vabs = []
vems = []
vabs_er = []
vems_er = []
for cx, cy in zip(xcen_e, ycen_e):
    popt, perr = mg_fit(cx,cy,z_e,plot=True)
    vabs.append(popt[0])
    vems.append(popt[1])
    vabs_er.append(perr[0])
    vems_er.append(perr[1])
coordinates = ['(114,226)','(114,233)','(115,237)','(141,264)','(148,269)']
#plt.errorbar(coordinates,vabs,yerr=vabs_er,fmt='o')
plt.errorbar(coordinates,vabs,yerr=vabs_er,fmt='o',label='Absorption')
plt.errorbar(coordinates,vems,yerr=vems_er,fmt='o',label='Emission')
plt.legend(loc=0)

plt.xlabel('Pixel Coordinates')
plt.ylabel('Velocity [km s$^{-1}$]')
plt.title('MgII Velocity for region E')
plt.savefig('/home/ahmed/astro/figures/fitting/regions/velocity_Mg_E.pdf',overwrite=True)
```
