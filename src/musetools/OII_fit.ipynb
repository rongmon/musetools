```python
from astropy.convolution import convolve, Gaussian1DKernel
from astropy.io import ascii
from astropy.table import Table
def model_OII(wave,z,tau,sigma):
    lam1 = 2470.97 #2324.21;
    lam1_obs = (1. + z) * lam1
    lam2 = 2471.09 #2325.40; 
    lam2_obs = (1. + z) * lam2
    c = 1. # Get the Doublet ratio
    F = 1. + c* tau * np.exp(- (wave - lam1_obs)**2. / (2. * sigma**2.)) + tau * np.exp(-(wave - lam2_obs)**2. / (2. * sigma**2.))
    muse_kernel = ((2.68871980676/1.25 )/ 2.355)#FWHM_avg = 2.50161030596 # Check the kernel for this part of the spectrum
    g = Gaussian1DKernel(stddev=muse_kernel)
    fmodel = convolve(F, g, boundary='extend')
    return fmodel
xcen = [114,114,115,118,121,124,127,133,137,141,148,153,244,240,238,242]
ycen = [226,233,237,241,244,248,252,257,260,264,269,271,237,234,228,224]
# xcen 160,166,170,177,185,191,198,203,208,213,220,225,231,238,244,246
# ycen 274,274,274,274,274,272,271,270,268,266,263,259,255,249,244,240
from astropy.wcs import WCS
w = WCS(header)
xcen_cord = []
ycen_cord = []
for cx, cy in zip(xcen, ycen):
    c=w.pixel_to_world_values(cx,cy,0)
    xcen_cord.append(c[0])
    ycen_cord.append(c[1])
z = []
z_er = []
for cx, cy in zip(xcen, ycen):
    data = ascii.read('/home/ahmed/astro/spectra/spectrum_OII_'+str(cx)+'_'+str(cy)+'.dat')
    wave = data[0][:]
    flx_norm = data[4][:]
    flx_er_norm = data[5][:]
    popt, pcov = curve_fit(model_OII, wave, flx_norm,[1.7,1.,50.],sigma=flx_er_norm,bounds=([1.7,0.,0.],[1.72,1.5,1000.]))
    perr = np.sqrt(np.diag(pcov))
    z.append(popt[0])
    z_er.append(perr[0])
    print('z: ',popt[0])
    print(popt)
    print(perr)
    fig, ax = plt.subplots()
    ax.step(wave,flx_norm,label='Normalized Flux')
    ax.step(wave,model_OII(wave,*popt),label='Model')
    ax.set_title('Fitting OII doublet for '+str(cx)+'_'+str(cy)+'')
    ax.set_xlabel('Wavelength')
    ax.set_ylabel('Normalized Flux')
    ax.set_xlim([6640.,6750.])
    ax.legend(loc=0, fontsize='small')
    #fig.savefig('/home/ahmed/astro/figures/fitting/Fit_CIII_'+str(cx)+'_'+str(cy)+'.pdf')
    plt.show()
info = Table([xcen, ycen, xcen_cord, ycen_cord, z, z_er], names = ['xpixel', 'ypixel', 'RA', 'Dec', 'redshift', 'error in redshift'])
ascii.write(info,'/home/ahmed/astro/spectra/redshiftOII.dat',overwrite=True)
```
